**Ejecución de script introductorio**

```python
print("Curso: Sistemas Inteligentes")
print("Sesión 1 - Introducción a IA")

# Ejemplo básico de operación matemática
a = 5
b = 3
suma = a + b
print("La suma es:", suma)

# Ejemplo de estructura condicional
if suma > 5:
    print("La suma es mayor que 5")
else:
    print("La suma es menor o igual que 5")
```

```python
print("Curso: Sistemas Inteligentes")
print("Sesión 1 - Introducción a IA")

# Operación matemática modificada
a = 127
b = 23.24
suma = a + b
print("La suma es:", suma)

# Ejemplo de estructura condicional
if suma > 5:
    print("La suma es mayor que 5")
else:
    print("La suma es menor o igual que 5")

# Entrada por teclado
nombre = input("¿Quién está ahí?")
edad = int(input("¿Cuántos años tienes?"))

if edad >= 18:
    frase = "¡Eres mayor de edad! :D"
elif 0 < edad < 18:
    frase = "Eres menor de edad :)"
print(f"Hola {nombre}. {frase}")
```
